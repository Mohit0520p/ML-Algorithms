```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('data_cleaned.csv')
df.head()
```

```python
df.shape
```

```python
#segregating the dependent and independent variables
x=df.drop(['Survived'],axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
#normalise the data because sklearn only works with {0,1}
```

```python
cols=train_x.columns
cols
```

```python
train_x_scaled=scaler.fit_transform(train_x)
train_x_scaled=pd.DataFrame(train_x_scaled,columns=cols)
train_x_scaled.head()
```

```python
test_x_scaled=scaler.fit_transform(test_x)
test_x_scaled=pd.DataFrame(test_x_scaled,columns=cols)
test_x_scaled.head()
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.metrics import f1_score
```

```python
logreg=LR()
logreg.fit(train_x,train_y)
train_predict=logreg.predict(train_x)
k=f1_score(train_predict,train_y)
print('Training F1_score',k)
```

```python
test_predict=logreg.predict(test_x)
k=f1_score(test_predict,test_y)
print('Testing F1_Score',k)
```

```python
#We can also find out the probabilites by using the predict_proba function
```

```python
train_predict=logreg.predict_proba(train_x)
train_predict
```

```python
train_pred=train_predict[:,1]
train_pred
```

```python
for i in range(len(train_pred)):
    if(train_pred[i]>0.55):
        train_pred[i]=1
    else:    
        train_pred[i]=0
```

```python
k=f1_score(train_pred,train_y)
print('Training F1_score',k)
```

```python
test_predict=logreg.predict_proba(test_x)
test_predict
```

```python
test_pred=test_predict[:,1]
test_pred
for i in range(len(test_pred)):
    if(test_pred[i]<0.5):
        test_pred[i]=1
    else:    
        test_pred[i]=0
```

```python
k=f1_score(test_pred,test_y)
print('Testing f1_score',k)
```

```python
from sklearn.metrics import confusion_matrix
cf=confusion_matrix(test_y,test_pred)
print(cf)
```

```python
#Parameters of logistic regression
logreg.coef_
```

```python
plt.figure(figsize=(10,15),dpi=50,facecolor='w',edgecolor='b')
x=range(len(train_x.columns))
c=logreg.coef_.reshape(-1)
plt.bar(x,c)
plt.xlabel('Variables')
plt.ylabel('Coefficients')
plt.title('Coefficient PLot')
plt.show()
```

```python
Coefficients=pd.DataFrame({'Variable':train_x.columns,'Coefficient':abs(c)})
Coefficients.head()
```

```python
high_var=Coefficients[Coefficients.Coefficient>0.5]
```

```python
subset=df[high_var['Variable'].values]
subset.head()
```

```python
train_x,test_x,train_y,test_y=train_test_split(subset,y,random_state=42)
```

```python
logreg=LR()
logreg.fit(train_x,train_y)
train_predict=logreg.predict(train_x)
k=f1_score(train_predict,train_y)
print('Training F1 score',k)
```

```python
test_predict=logreg.predict(test_x)
k=f1_score(test_predict,test_y)
print('Test F1_score',k)
```

```python
plt.figure(figsize=(8,6),dpi=50,facecolor='w',edgecolor='b')
x=range(len(train_x.columns))
y=logreg.coef_.reshape(-1)
plt.bar(x,y)
plt.xlabel('Variables')
plt.ylabel('Coefficients')
plt.title('Coefficient plot')
plt.show()
```

```python
#as we can observe , the f1_score has gone down but our model is only left with variables with utmost significance.
```
